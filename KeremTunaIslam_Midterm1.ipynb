```python
#QUESTION 1 

import random as mt
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

L = 200 #set our list length
mylist= [] #define empty list

#generate a list of L length containing random integers between -9 and 9
for x in range(0,L): 
  mylist.append(mt.randint(-9,9))

movingAvg = [] #define empty list

#moving average taking loop
for item_number in range(1, len(mylist)+1): 
  avg = 0 #set average to zero
  for i in range(item_number, 0, -1):
      avg = avg + mylist[i-1] 
   
    # rounds moving averages the the 4th decimal
  movingAvg.append(round(avg/len(mylist),4))

#print the outputs
print('My First List:')    
print(mylist)
print("\n\n")
print('Moving Average List:')
print(movingAvg)
print("\n\n")
print("\n\n")
#plot random list
#plt.plot(mylist)
print('Moving Average Plot:')
#plot moving average
plt.plot(movingAvg)
```
